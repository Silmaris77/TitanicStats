```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

os.chdir(r"C:\Projects\TitanicStats")
data = pd.read_csv('titanic.csv')
os.getcwd()


# Podstawowe statystyki opisowe zmiennej 'Fare'
fare_stats = data['Fare'].describe().round(2)
print(f"{fare_stats}\n")

# Obliczenie kwartylów, średniej i mediany
fare_mean = round(data['Fare'].mean(),2)
fare_median = round(data['Fare'].median(),2)
fare_mode = stats.mode(data['Fare'])

q1 = round(np.quantile(data['Fare'], 0.25),2)
q3 = round(np.quantile(data['Fare'], 0.75),2)

#obliczenie zakresu, wariancji, odchylenia standardowego i skośności
range_fare = round(max(data['Fare']) - min(data['Fare']),2)
variance_fare = round(np.var(data['Fare'], ddof=1),2)
fare_std = round(data['Fare'].std(),2)
fare_skew = round(stats.skew(data['Fare']),2)


print("Miary położenia:")
print(f"Średnia: {fare_mean}\nMediana: {fare_median}\nModa: {fare_mode}\nPierwszy kwartyl (Q1): {q1}\nTrzeci kwartyl (Q3): {q3}\n")

print("Miary rozproszenia:")
print(f"Zakres: {range_fare}\nWariancja: {variance_fare}\nOdchylenie standardowe: {fare_std}\nSkośność: {fare_skew}\n")



# Wizualizacje
plt.figure(figsize=(10, 3))
sns.histplot(data['Fare'], bins=40, kde=True)
plt.title('Histogram zmiennej Fare')
plt.xlabel('Wysokość opłaty (Fare)')
plt.ylabel('Liczba biletów')
plt.axvline(x = fare_mean, color = 'red', label = 'Średnia')
plt.axvline(x = fare_median, color = 'green', label = 'Mediana')
plt.axvline(x = fare_mode[0], color = 'orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(10, 3))
sns.boxplot(x=data['Fare'])
plt.title('Wykres pudełkowy zmiennej Fare')
plt.xlabel('Wysokość opłaty (Fare)')
plt.show()
```
